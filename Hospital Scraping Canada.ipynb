```python
import numpy as np
def getName():
    try:
        try:
            driver.find_element(By.CLASS_NAME, 'sub-headline').text == "We can't seem to find the page you are looking for."
            return
        
        except NoSuchElementException:
            try:
                info_cont = driver.find_element(By.CLASS_NAME, 'split-50-50')
                name = info_cont.find_element(By.TAG_NAME, 'h1').text
                phone_number=info_cont.find_element(By.TAG_NAME, 'a').text
                fax_number = np.nan
                
                main_container = driver.find_element(By.CLASS_NAME,'split-50-50')
                address_info = main_container.find_element(By.CLASS_NAME,'contact-info')
                address = address_info.text.split('\n')[:-1]
                hours_lst = []
                try:
                    hours = driver.find_element(By.XPATH, '//*[@id="Main"]/section[2]/div[1]/section/div/div[3]')
                except NoSuchElementException:
                    hours = driver.find_element(By.XPATH, '//*[@id="Main"]/section[3]/div[1]/section/div/div[3]')
                
                for hour in hours.find_elements(By.TAG_NAME, 'li'):
                    hours_lst.append(hour.text.split('\n'))
                    
                return name, phone_number, fax_number, address, hours_lst    
            
            except NoSuchElementException:
                print('error found')
                #NOTE Designed according to page 3 Calgary West Animal Hospital:
                main_container = driver.find_element(By.CLASS_NAME, 'sh-contact-map__txt-block')
                name=main_container.find_element(By.CLASS_NAME, 'sh-contact-map__hospital').text
                text_contatiner = main_container.find_element(By.CLASS_NAME,'sh-contact-map__txt-indent').find_elements(By.TAG_NAME,'p')
                phone_number = text_contatiner[-2].text.split(' ')[-1].strip()
                fax_number = text_contatiner[-1].text.split(' ')[-1].strip()
                
                main_container = driver.find_element(By.CLASS_NAME, 'sh-contact-map__txt-block')
                
                text_contatiner = main_container.find_element(By.CLASS_NAME,'sh-contact-map__txt-indent').find_elements(By.TAG_NAME,'p')
                address_1 = text_contatiner[0].text.split('\n')
                address_2 = text_contatiner[1].text.split('\n')
                
                final_address = address_1 + address_2
                hours = main_container.find_element(By.TAG_NAME,"ul").text    
                
                return name, phone_number, fax_number, final_address, hours
            
    except NoSuchElementException:
        name = driver.title
        try:
            main_section = driver.find_element(By.CLASS_NAME, 'col-2-green')
        except NoSuchElementException:
            main_section = driver.find_element(By.CLASS_NAME, 'section-col-2')
        address = main_section.find_element(By.CLASS_NAME,'adresse')
        address = address.text.split('\n')[:-1]
        phone_number = main_section.find_element(By.CLASS_NAME,'tel').text
        fax_number = np.nan
        print(address)
        hours = np.nan
        return name, phone_number, fax_number, address,hours

                        
```

```python
def extract_director():
    director_name = None  
    director_img_link = None
    try:
        director_img_link = driver.find_element(By.ID,'staffType_Administration').find_element(By.TAG_NAME, 'img').get_attribute('src')
        director_name=driver.find_element(By.ID,'staffType_Administration').find_element(By.CLASS_NAME,'content').find_element(By.TAG_NAME,'h3').text
        print(director_name)
        print(director_img_link)
        information_dict['Name'].append(director_name)
        information_dict['Image Link'].append(director_img_link)
        return
    except NoSuchElementException:
        try:
            try:
                director_img_link_cont=driver.find_elements(By.CLASS_NAME, 'sh-rotating-bios__vet')
                director_img_link = director_img_link_cont[0].find_element(By.TAG_NAME, 'img').get_attribute('src')
            except IndexError:
                director_img_link = None
                name_cont = None
                print(director_name)
                print(director_img_link)
                information_dict['Name'].append(director_name)
                information_dict['Image Link'].append(director_img_link)
                return    
            name_cont=driver.find_elements(By.CLASS_NAME, 'sh-rotating-bios__vet')[0]
            director_name=name_cont.find_element(By.CLASS_NAME, 'sh-rotating-bios__vet-name').find_element(By.TAG_NAME, 'span').text    
            print(director_name)
            print(director_img_link)
            information_dict['Name'].append(director_name)
            information_dict['Image Link'].append(director_img_link)
            return
        except IndexError or NoSuchElementException:
            director_img_link = None
            name_cont = None
            print(director_name)
            print(director_img_link)
            information_dict['Name'].append(director_name)
            information_dict['Image Link'].append(director_img_link)
            return



# for city, links in all_links_dict.items():
#     iteration = 0
#     for link,clinic_name in links:
#         print(clinic_name,city,iteration)
#         information_dict['Clinic Name'].append(clinic_name)
#         driver.get(link + '/team')
#         time.sleep(random.randint(1,5))
#         extract_director()
#         iteration += 1
```

```python
for city, links in all_links_dict.items():
    count = 0 
    for link, clinic_name in links:
        
        driver.get(link)
        time.sleep(random.randint(1, 3))
        #i am printing the link for debugging 
        if driver.title in ['Hôpitaux Vétérinaires 24h | Vet et Nous','Vet et Nous: Vétérinaire de Confiance pour le Soin de vos Animaux']:
            continue
        count += 1
        results = getName()
        
        # Check if results is not None before accessing its elements
        if results is not None:
            # Ensure that all elements in results are not None before appending to information_dict
            if all(elem is not None for elem in results):
                information_dict['City Name'].append(city)
                information_dict['Clinic Name'].append(results[0])
                information_dict['Phone Number'].append(results[1])
                information_dict['Fax'].append(results[2])
                information_dict['Address'].append(results[3])
                information_dict['Hours'].append(results[4])
                print(results)
                driver.get(link + '/team')
                time.sleep(1)
                extract_director()
                information_dict['Website Link'].append(link)
            else:
                print("Some elements in results are None:", results)
        else:
            print("Results is None.")
```

```python
lstreet =[]
lcity =[]
lstate =[]
lzip_code = []

def field_seperator(check):
    parts =check[0].split(',')
    print(len(parts),'->', parts)
    if len(parts) == 4:
        street, city, state, zip_code=parts[0],parts[1],parts[2],parts[3]        
        lstreet.append(street)
        lcity.append(city)
        lstate.append(state)
        lzip_code.append(zip_code) 
        return street, city, state, zip_code
    
    elif len(parts) == 5:
        street, city, state, zip_code= parts[0] + parts[1], parts[2],parts[3], parts[4]        
        lstreet.append(street)
        lcity.append(city)
        lstate.append(state)
        lzip_code.append(zip_code)
        return street, city, state, zip_code
    
    elif len(parts) == 1:
        if parts[0] == '':
            street, city, state, zip_code= np.nan, np.nan, np.nan, np.nan    
            lstreet.append(street)
            lcity.append(city)
            lstate.append(state)
            lzip_code.append(zip_code)
            return street, city, state, zip_code

        street, city, state, zip_code=parts[0], np.nan, np.nan, np.nan         
        lstreet.append(street)
        lcity.append(city)
        lstate.append(state)
        lzip_code.append(zip_code)
        return street, city, state, zip_code
    
for address in df['Address']:
    print('Answer:',field_seperator(address))
```
