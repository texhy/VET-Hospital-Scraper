```python
#THIS IS NOTE FOR THE ONE WITH NO BOX CONTAINER
iter = 0

def extract_directors(link):
    global iter 
    director_name = None
    img_link = None
    try:
        try:
            container = driver.find_element(By.CLASS_NAME, 'team-spotlight')
            # slide = container.find_elements(By.CLASS_NAME, 'slick-slide')
            # memeber_info_cont = slide[0].find_element(By.ID,'team-member-info')
            director_name = container.find_element(By.CLASS_NAME,'team-spotlight__name').text
            memeber_media_cont = container.find_element(By.ID,'team-member-media')
            img=memeber_media_cont.find_element(By.TAG_NAME, 'img')
            img_link = img.get_attribute('src')
            print(director_name)
            print(img_link)
            information_dict['Name'].append(director_name)
            information_dict['Image Link'].append(img_link)
            return
            
        except NoSuchElementException:
            service_slider_container = driver.find_element(By.CLASS_NAME,'service-slider')
            memebers_container = service_slider_container.find_elements(By.CLASS_NAME,'service-slider__item')
            #NOTE we will always take the first container as it has the director information
            director_name = memebers_container[0].find_element(By.CLASS_NAME, 'service-slider__item__name').text
            director_image = memebers_container[0].find_element(By.CLASS_NAME, 'service-slider__item__image').find_element(By.TAG_NAME,'img').get_attribute('src')
            print(director_name)
            print(director_image)
            information_dict['Name'].append(director_name)
            information_dict['Image Link'].append(director_image)
            

            return
    except NoSuchElementException:
        try:
            director_name = driver.find_element(By.XPATH, '/html/body/div[5]/div/div[1]/p').text
            image = None
            print(director_name.split(',')[0])
            print(image)
            information_dict['Name'].append(director_name)
            information_dict['Image Link'].append(image)
            
            return
        except NoSuchElementException:
            iter += 1
            if iter <= 1:
                driver.get(link)
                extract_directors(link)
            return            
```

```python
def extract_information(city_name):
    information_dict['City Name'].append(city_name)
    #on the data extraction page
    cont = driver.find_element(By.CLASS_NAME, 'hospital-info')
    #getting the name of the clinic
    #the upper container has the names and animals we see info

    upper_cont = cont.find_element(By.CLASS_NAME, 'hospital-info__titles')
    name_clinic = upper_cont.find_element(By.CLASS_NAME,'hospital-info__title').text
    animals_checked = upper_cont.find_element(By.CLASS_NAME, 'hospital-info__animal-list')
    animals = ' '.join(animals_checked.text.split()[3:])
    information_dict['Clinic Name'].append(name_clinic)
    information_dict['Animals'].append(animals)
    #-----------------------------------#
    try:
        lower_cont = cont.find_element(By.XPATH, '/html/body/section[3]/div/div/div/div[2]')
    except NoSuchElementException:
        lower_cont = cont.find_element(By.CLASS_NAME, 'hospital-info__middle ')
        
    list_of_info = lower_cont.find_elements(By.TAG_NAME, 'li')
    #getting the contact number:
    try:
        phone_number = list_of_info[0].text.split('\n')[1]
        fax_number = list_of_info[0].text.split('\n')[2]
        if fax_number == 'Press Inquiries >'or fax_number == 'Contact' or phone_number == 'Contact':
            fax_number == 'NULL'
            phone_number == 'NULL'
            
    except IndexError:
        phone_number = "NULL"
        fax_number = "NULL"
        
    information_dict['Phone Number'].append(phone_number)
    information_dict['Fax'].append(fax_number)

    #getting the location

    try:
        try:
            address = list_of_info[1].text.split('\n')[1:3]
                
            print('Number of elem is 3|Container Loc at 1: ',address)
        except IndexError:
            address = list_of_info[1].text.split('\n')[1:4]
            print('Number of elem is 4|Container Loc at 1: ',address)    
        
        if 'Days Hours' in address:
            raise IndexError
        
        try:
            hours=list_of_info[2].find_element(By.ID,'accordion-hospital-hours').find_element(By.CLASS_NAME,'accordion__item').text.split('\n')[1:]
            hours_string = ' '.join(hours)
            information_dict['Hours'].append(hours_string)
        
        except NoSuchElementException:
            hours = list_of_info[1].text.split('\n')[1:]
            hours_string = ' '.join(hours)
            information_dict['Hours'].append(hours_string)
            
    except IndexError:
        try:
            address = list_of_info[2].text.split('\n')[1:3]
            print('Number of elem is 3|Container Loc at 2: ',address)
        except IndexError:
            address = list_of_info[2].text.split('\n')[1:4]
            print('Number of elem is 4|Container Loc at 2: ',address)
        
        hours=list_of_info[1].text.split('\n')[2:]
    
    hours_string = ' '.join(hours)    
    information_dict['Address'].append(address)
    information_dict['Hours'].append(hours_string)
    

    
    # location #'Location 719 Quintard Avenue Anniston, AL 36201 Get directions'
    #NOTE i need to remove ''Location and Get Directions when looping through
    print("**************")
    print(city_name)
    print(name_clinic)
    print(animals)
    print(phone_number)
    print(fax_number)
    print(address)
    print(hours_string)
    print("**************")
    
```
